```python
import pandas as pd
```

```python
df = pd.read_csv("tcs_only_modeling_abt.tsv", sep='\t')
```

```python
data = df.drop(["tech_obs_id", "app_bank_nm", "tech_dog_type", "trg_utilization", "tech_target_source", "tech_target_d"],
               axis = 1)
data = data.drop(['var_0{}'.format(i) for i in range(63, 73)], axis = 1)
data = data.dropna()
data = pd.get_dummies(data)
data.head()
```

```python
tf = pd.read_csv('default_grace_scores.csv')
tf.head()
```

```python
scores = data.join(tf.set_index('uid'), on='uid')
```

```python
scores.dropna(inplace=True)
scores.drop(columns=['financial_application_rk', 'retro_date'], inplace=True)
scores.head()
```

```python
logit_pd = scores['logit_default']
logit_gr = scores['logit_grace03']
scores.drop(columns=['logit_default', 'logit_grace03'], inplace=True)
```

```python
scores.set_index('uid', inplace=True)
```

```python
scores.head()
```

```python
y_pd = scores['trg_pd']
y_gr = scores['trg_grace']
scores.drop(columns=['trg_pd', 'trg_grace'], inplace=True)
scores.head()
```

```python
features = {'be' : ['var_001',
 'var_039',
 'var_046',
 'var_049',
 'var_051',
 'var_078',
 'var_084',
 'var_090',
 'var_091',
 'var_107',
 'var_123',
 'var_133',
 'var_136',
 'var_138',
 'var_175',
 'var_182',
 'var_188',
 'var_189',
 'var_198',
 'var_203',
 'var_220',
 'var_224',
 'var_234',
 'var_242',
 'var_254',
 'var_257',
 'var_286',
 'var_292',
 'var_312',
 'var_338',
 'var_365',
 'var_373',
 'var_378',
 'var_390',
 'var_398',
 'var_405',
 'var_406',
 'var_429',
 'var_440',
 'var_446',
 'var_447',
 'var_461',
 'var_062_orient']}
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
```

```python
X_train, X_test, y_train, y_test, log_train, log_test = train_test_split(scores, y_pd, logit_pd, test_size=0.2, stratify=y_pd)
```

```python
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

```python
clf = LogisticRegressionCV(cv=5, 
                           random_state=0, 
                           scoring='neg_log_loss', 
                           verbose=1).fit(X_train[features['be']], y_train)
```

```python
logs = clf.predict_log_proba(X_train[features['be']])
```

```python
import numpy as np
```

```python
def make_logit(logs):
    return -np.log(logs[..., 1] / logs[..., 0])
```

```python
logits_train = make_logit(logs)
```

```python
scores_logs_train = pd.DataFrame({'my_scores' : logits_train, 'tf_scores' : log_train})
```

```python
scores_logs_train.head()
```

```python
combined_clf = LogisticRegressionCV(cv=5, 
                           random_state=0, 
                           scoring='neg_log_loss', 
                           verbose=1).fit(scores_logs_train, y_train)
```

```python
logits_test = make_logit(clf.predict_log_proba(X_test[features['be']]))
scores_logs_test = pd.DataFrame({'my_scores' : logits_test, 'tf_scores' : log_test})
scores_logs_test.head()
```

```python
my_final_loss = combined_clf.score(scores_logs_test, y_test)
```

```python
my_final_loss
```

```python
probs = np.exp(log_test) / (1 + np.exp(log_test))
```

```python
tf_final_loss = np.sum( - y_test * np.log(probs) - (1 - y_test) * np.log(1 - probs))
```

```python
tf_final_loss
```
