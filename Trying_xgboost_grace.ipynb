```python
import xgboost as xg
```

```python
import pandas as pd
```

```python
df = pd.read_csv("tcs_only_modeling_abt.tsv", sep='\t')
data = df.drop(["uid", "tech_obs_id", "app_bank_nm", "tech_dog_type", "trg_utilization", "tech_target_source", "tech_target_d"],
               axis = 1)
data.head()
```

```python
data = data.drop(['var_0{}'.format(i) for i in range(63, 73)], axis = 1)
data = data.dropna()
data = pd.get_dummies(data)
data.head()
```

```python
y_pd = data['trg_pd']
y_gr = data['trg_grace']
data = data.drop(['trg_grace', 'trg_pd'], axis=1)
data.head()
```

# Пробуем xgboost на всех признаках

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
data = pd.DataFrame(scaler.fit_transform(data), columns = data.columns)
```

```python
PD_train, PD_test, y_pd_train, y_pd_test = train_test_split(data, y_pd, stratify=y_pd)
```

```python
GR_train, GR_test, y_gr_train, y_gr_test = train_test_split(data, y_gr, stratify=y_gr)
```

```python
Train_gr = xg.DMatrix(GR_train, label=y_gr_train)
```

```python
Test_gr = xg.DMatrix(GR_test, label=y_gr_test)
```

```python
params = {'objective' : 'binary:logistic'}
params['eval_metric'] = ['error', 'auc']
params['max_depth'] = 5
params['silent'] = False
params['scale_pos_weight'] = 1
params['learning_rate'] = 0.1
params['colsample_bytree'] = 0.4
params['subsample'] = 1
params['n_estimators'] = 100
params['reg_alpha'] = 0.3
params['gamma'] = 5
```

```python
evallist = [(Test_gr, 'eval'), (Train_gr, 'train')]
```

```python
bst = xg.train(params, Train_gr, evals=evallist, early_stopping_rounds=50)
```

```python
%matplotlib inline
```

```python
dic = bst.get_score(importance_type='gain')
```

```python
import operator
```

```python
features_sorted = sorted(dic.items(), key=operator.itemgetter(1), reverse=True)
```

```python
import numpy as np
```

```python
np.array(features_sorted[:17])[..., 0]
```

```python
from catboost import CatBoostClassifier, Pool
```

```python
train_pool = Pool(GR_train, y_gr_train)
test_pool = Pool(GR_test, y_gr_test)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
model = CatBoostClassifier(iterations=10,
                           depth=4,
                           learning_rate=1,
                           loss_function='Logloss',
                           verbose=True)
# train the model
model.fit(train_pool)
# make the prediction using the resulting model
preds_proba = model.predict_proba(test_pool)
print(roc_auc_score(y_gr_test, preds_proba[..., 1]))
```
